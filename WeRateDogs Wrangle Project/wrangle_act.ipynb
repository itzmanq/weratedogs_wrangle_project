#### df_image_clean
#### Define
 - Create two columns (dog_breed and prediction_confidence) by merging p1, p2, and p3 as well as p1_conf, p2_conf and p3_conf
 - Merge dog_breed and p_confidence columns with df_archive_clean table  

#### Code

```python
p_dog = [(df_image_clean['p1_dog'] == True), (df_image_clean['p2_dog'] == True), (df_image_clean['p3_dog'] == True)]

dog_breed = [df_image_clean['p1'], df_image_clean['p2'], df_image_clean['p3']]

p_confidence = [df_image_clean['p1_conf'], df_image_clean['p2_conf'], df_image_clean['p3_conf']]

# select dog_breed based on if conditions are met
df_image_clean['dog_breed'] = np.select(p_dog, dog_breed, 
                                       default = 'none')

# select the p_confidence based on if conditions are met
df_image_clean['p_confidence'] = np.select(p_dog, p_confidence, 
                                            default = 0)
```

```python
df_image_clean.drop(df_image_clean[df_image_clean['dog_breed'] == 'none'].index, inplace = True)
```

```python
# Select columns to be merged with df_archive_clean table
prediction = df_image_clean[['tweet_id', 'dog_breed', 'p_confidence']]
```

```python
# Merge prediction table with df_archive_clean table
df_archive_clean = df_archive_clean.merge(prediction, how= 'left', on= 'tweet_id')
```

```python
# fill the missing values in column dog_breed with 'None' and p_confidence with 0
df_archive_clean['dog_breed'] = df_archive_clean['dog_breed'].fillna(value='None')
df_archive_clean['p_confidence'] = df_archive_clean['p_confidence'].fillna(value= 0)
```
